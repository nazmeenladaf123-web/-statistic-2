Question 1: What is the difference between descriptive statistics and inferential Statistics? Explain with examples.
 Ans :- Descriptive statistics and inferential statistics are two important branches of statistics that serve different purposes. Descriptive statistics focus on organizing, summarizing, and presenting data in a meaningful way. They help us understand the main features of a dataset by using measures such as mean, median, mode, standard deviation, percentages, tables, and graphs. For example, if a teacher collects the marks of 50 students and calculates the average score, highest score, and lowest score, this is an application of descriptive statistics because it describes the collected data. On the other hand, inferential statistics are used to make predictions, estimates, or conclusions about a larger population based on a sample of data. It uses techniques such as hypothesis testing, confidence intervals, and regression analysis. For example, if a researcher surveys 1,000 people to estimate the average income of all residents in a city, the researcher is using inferential statistics because conclusions about the entire population are being drawn from a sample. In summary, descriptive statistics describe and summarize existing data, whereas inferential statistics help in making decisions and generalizations about a larger population.


Question 2: What is sampling in statistics? Explain the differences between random and stratified sampling.

Ans :-  Sampling in statistics is the process of selecting a subset of individuals, observations, or data points from a larger population to study and draw conclusions about the entire population. Sampling is important because it saves time, cost, and effort while still providing reliable information about the population.

There are different methods of sampling, among which random sampling and stratified sampling are widely used. In random sampling, every member of the population has an equal chance of being selected, ensuring fairness and reducing bias. For example, selecting 100 students randomly from a university using a lottery method is an example of random sampling.

 In contrast, stratified sampling involves dividing the population into distinct groups or strata based on specific characteristics such as age, gender, or income level, and then selecting samples from each group. This method ensures that all important subgroups of the population are represented. For example, if a school has students from different grades, researchers may divide the students by grade level and randomly select participants from each grade.

  The main difference between the two methods is that random sampling treats the entire population as one group, whereas stratified sampling divides the population into smaller groups before selecting samples, leading to better representation and often more accurate results.

Question 3: Define mean, median, and mode. Explain why these measures of central tendency are important.

Mean, median, and mode are the three primary measures of central tendency used in statistics to describe the center or typical value of a dataset. The mean, also known as the average, is calculated by adding all the values in a dataset and dividing the total by the number of observations. For example, the mean of the numbers 2, 4, 6, and 8 is (2 + 4 + 6 + 8) ÷ 4 = 5. The median is the middle value of a dataset when the values are arranged in ascending or descending order. If there is an even number of observations, the median is the average of the two middle values. The mode is the value that occurs most frequently in a dataset. These measures of central tendency are important because they help summarize large amounts of data into a single representative value, making the data easier to understand and interpret. The mean provides an overall average, the median is useful when the data contains outliers or is skewed, and the mode helps identify the most common value or category. Together, these measures enable researchers, analysts, and decision-makers to understand patterns, compare datasets, and make informed decisions based on the data.

Question 4: Explain skewness and kurtosis. What does a positive skew imply about the data?

Ans :- Skewness and kurtosis are statistical measures used to describe the shape of a data distribution. Skewness measures the degree of asymmetry in a distribution around its mean. A distribution is considered symmetric if the left and right sides are mirror images of each other. If the distribution is not symmetric, it is said to be skewed. Kurtosis, on the other hand, measures the peakedness and the heaviness of the tails of a distribution compared to a normal distribution. High kurtosis indicates that the data have heavier tails and a sharper peak, while low kurtosis indicates lighter tails and a flatter distribution. A positive skew (also called right skewness) implies that the distribution has a longer tail on the right side, meaning that a few unusually high values pull the mean toward the right. In a positively skewed distribution, the mean is typically greater than the median. For example, income distributions are often positively skewed because a small number of individuals earn significantly higher incomes than the majority of the population. Understanding skewness and kurtosis is important because they help analysts assess the distribution of data, identify outliers, and choose appropriate statistical methods for analysis

```python
import statistics

# Given list of numbers
numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

# Calculate mean, median, and mode
mean_value = statistics.mean(numbers)
median_value = statistics.median(numbers)
mode_value = statistics.multimode(numbers)  # Handles multiple modes

# Display the results
print("Numbers:", numbers)
print("Mean:", mean_value)
print("Median:", median_value)
print("Mode:", mode_value)
```

```python
import numpy as np

# Given datasets
list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

# Calculate covariance matrix
covariance_matrix = np.cov(list_x, list_y)

# Calculate correlation coefficient matrix
correlation_matrix = np.corrcoef(list_x, list_y)

# Display results
print("Covariance Matrix:")
print(covariance_matrix)

print("\nCorrelation Coefficient Matrix:")
print(correlation_matrix)

print("\nCovariance:", covariance_matrix[0,1])
print("Correlation Coefficient:", correlation_matrix[0,1])
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Given data
data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

# Calculate quartiles and IQR
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1

# Calculate outlier boundaries
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Find outliers
outliers = [x for x in data if x < lower_bound or x > upper_bound]

# Display results
print("Q1 =", Q1)
print("Q3 =", Q3)
print("IQR =", IQR)
print("Lower Bound =", lower_bound)
print("Upper Bound =", upper_bound)
print("Outliers =", outliers)

# Draw boxplot
plt.figure(figsize=(8, 4))
plt.boxplot(data)
plt.title("Boxplot of Numeric Data")
plt.ylabel("Values")
plt.show()
```

```python
import numpy as np

# Given data
advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

# Calculate covariance matrix
covariance = np.cov(advertising_spend, daily_sales)

# Calculate correlation coefficient matrix
correlation = np.corrcoef(advertising_spend, daily_sales)

# Display results
print("Covariance Matrix:")
print(covariance)

print("\nCorrelation Matrix:")
print(correlation)

print("\nCovariance:", covariance[0, 1])
print("Correlation Coefficient:", correlation[0, 1])
```

```python
import matplotlib.pyplot as plt
import statistics

# Survey data
survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

# Calculate summary statistics
mean_score = statistics.mean(survey_scores)
median_score = statistics.median(survey_scores)
mode_score = statistics.mode(survey_scores)
std_dev = statistics.stdev(survey_scores)

# Display summary statistics
print("Mean:", mean_score)
print("Median:", median_score)
print("Mode:", mode_score)
print("Standard Deviation:", round(std_dev, 2))
print("Minimum:", min(survey_scores))
print("Maximum:", max(survey_scores))

# Create histogram
plt.figure(figsize=(8, 5))
plt.hist(survey_scores, bins=7, edgecolor='black')

# Add labels and title
plt.title("Customer Satisfaction Survey Scores")
plt.xlabel("Survey Score")
plt.ylabel("Frequency")

# Display the plot
plt.show()
```
